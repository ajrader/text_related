<h3> Separability Statistics</h3>
<ol>
<li> $d_m$ number matching</li>
<li> $f_m$ frequency matching</li>
<li> $n_x$ mismatching arity</li>
<li> $f_x$ frequency mismatching</li>
</ol>

```python
#define matching function statistics
from collections import defaultdict
def createRefcounts(dataset):
    # requires dataset as input; generates defaultdict as output
    featureDict = defaultdict(int)
    for i in xrange(0,len(dataset)):
        for j in xrange(0,len(dataset[i])):
            featureDict[dset[i][j]] += 1
    
    return featureDict, len(dset)    
```

```python
featDict,nitems = createRefcounts(dset)
print featDict
nitems
```

```python
sum(map(lambda x: featDict[x]*featDict[x]/float(len(dset)), z))
```

```python
#Function to count matching statistics
def matchingMetrics(referenceDict, nitems,x):
    # given a reference dataset as a dictionary and an instance, x calculate d_m and f_m
    dm = float(sum(map(lambda element: referenceDict[element], x)))/float(nitems)
    fm = float(sum(map(lambda element: referenceDict[element]**2/float(nitems), x)))
    return dm, fm
```

```python
print matchingMetrics(featDict,10,z)
matchingMetrics(featDict,10,y)
```
