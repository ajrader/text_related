```python
try:
    from xml.etree.cElementTree import XML
except ImportError:
    from xml.etree.ElementTree import XML
import zipfile


 
"""
Module that extract text from MS XML Word document (.docx).
(Inspired by python-docx <https://github.com/mikemaccana/python-docx>)
"""
 
WORD_NAMESPACE = '{http://schemas.openxmlformats.org/wordprocessingml/2006/main}'
PARA = WORD_NAMESPACE + 'p'
TEXT = WORD_NAMESPACE + 't'
 
def get_docx_soup(path):
    with zipfile.ZipFile(path) as document:
        xml_content = document.read('word/document.xml')
    soup = BeautifulSoup(xml_content)
    return soup
    
    
def get_docx_etree(path):
    """
    Take the path of a docx file as argument, return the text in unicode.
    """
    with zipfile.ZipFile(path) as document:
        xml_content = document.read('word/document.xml')
    
    tree = XML(xml_content)
    return tree
    
def get_docx_text(path):
    """
    Take the path of a docx file as argument, return the text in unicode.
    """
    tree = get_docx_etree(path)
 
    paragraphs = []
    for paragraph in tree.getiterator(PARA):
        texts = [node.text.encode('ascii','ignore')
                 for node in paragraph.getiterator(TEXT)
                 if node.text]
        if texts:
            paragraphs.append(''.join(texts))
 
    return '\n\n'.join(paragraphs)

def get_docx_text_from_tree(tree):
    """
    Take the path of a docx file as argument, return the text in unicode.
    """
    paragraphs = []
    for paragraph in tree.getiterator(PARA):
        texts = [node.text.encode('ascii','ignore')
                 for node in paragraph.getiterator(TEXT)
                 if node.text]
        if texts:
            paragraphs.append(''.join(texts))
 
    return '\n\n'.join(paragraphs)


```

```python
import re
def heading(tag):
    return tag.name == 'w:p' and len(tag('w:top',attrs={'w:color':"C0504D"})) > 0

def groups(tag):
    return tag.name == 'w:p' and len(tag('w:b')) > 0 and len(tag('w:t')) > 0  and not heading(tag)

def teams(tag):
    return tag.name == 'w:p' and len(tag('w:ind',attrs={'w:left':True})) > 0 and len(tag('w:t', text=re.compile("s team:",re.I|re.U))) > 0

def raw_txt(tag):
    return tag.name == 'w:p' and len(tag('w:t')) > 0 and not heading(tag) and not groups(tag) and not teams(tag)
```

```python
import pandas as pd

def heading_date(heading):
    hstr = heading.get_text().encode('ascii','ignore')
    matchObj = re.match( r'Weekly Staff Notes\s+(\S+\s?\S+,\s\d{4})', hstr,re.I)
    return matchObj.group(1)

def parse_soup(soup, df=None):
    if df is None: df = pd.DataFrame(columns=['Date','Group','Team','Text'])
    date = heading_date(soup.find(heading))

    for tag in soup(raw_txt):
        group = tag.find_previous(groups)
        group_txt = group.get_text()

        team = tag.findPreviousSibling(teams)
        if team:
            prevGroup = team.findPreviousSibling(groups)
            if prevGroup != group: team = None

        if team is not None: team_txt = team.get_text()
        else: team_txt = None

        df = df.append({'Date':date,'Group':group_txt,'Team':team_txt,'Text':tag.get_text()}, ignore_index=True)

    return df
```

## create a version that removes just the standard stop words ('english') and create a tfidf on this

```python
import nltk as nltk
nltk.data.path.append('/data/san/kes1/public_data/nltk_data')
from nltk.stem import SnowballStemmer
from nltk.tokenize.punkt import PunktWordTokenizer
from nltk.stem import WordNetLemmatizer

stemmer = SnowballStemmer('english')
punkt_tokenizer = PunktWordTokenizer()
wordnet_lemmatizer = WordNetLemmatizer()

def stem_tokenizer(x):
    return [stemmer.stem(s) for s in punkt_tokenizer.tokenize(x)]

def lemm_tokenizer(x):
     return [wordnet_lemmatizer.lemmatize(s) for s in punkt_tokenizer.tokenize(x)]
    
    
from sklearn.feature_extraction.text import _check_stop_list
stoplist = set(_check_stop_list('english'))
```

```python
from itertools import chain
def createTokens(list_of_text,mechanism='stem',stopwords=stoplist,remove_once = False):
    """
    inputs:
        list_of_text is a list of lists where length is the number of 'bullet points'
        mechanism is the tokenizer to use (stemmer or lemmatizer)
        stoplist is the stoplist to use
    output:
        a 
    """
    nmin = -1
    letterSet = set('qwertyuiopasdfghjklz xcvbnm/-`&')
    if mechanism == 'stem':
        texts = [[stemmer.stem(word) for word in (''.join(x for x in document.lower() if x in letterSet)).split() 
          if word not in stopwords] for document in list_of_text]
    elif mechanism== 'lemma':
        texts = [[wordnet_lemmatizer.lemmatize(word) for word in (''.join(x for x in document.lower() if x in letterSet)).split() 
          if word not in stopwords] for document in list_of_text]
    else:
        print "sorry that mechanism is not yet implemented"
        return
    
    #all_tokens = sum(texts)#, [])
    all_tokens = list(chain.from_iterable(texts))
    if remove_once:
        tokens_once = set(word for word in set(all_tokens) if all_tokens.count(word)< (0.01 * len(texts)))
        my_text = [' '.join([word for word in text if word not in tokens_once]) for text in texts if len(text)>5]
    
    else:
        my_text = [' '.join([word for word in text]) for text in texts if len(text)>nmin]

    my_words = list(chain.from_iterable([line.split(' ') for line in my_text]))    
    return(my_text,my_words)
```

## Describe/Determine what each group represents


```python
from sklearn.ensemble import RandomForestClassifier
def plotGroupDetails(X,cluster,terms,nestimators=100):
    nclusters = cluster.n_clusters
    X_dense = X.toarray()
    y = cluster.labels_
    obs_per_cluster = pd.Series(cluster.labels_).value_counts()
    for k in range(nclusters):
        y_binary = pd.Series(y).replace(k, nestimators)
        y_binary = y_binary.replace([e for e in range(nclusters) if e != k], 0)
        Forest = RandomForestClassifier(n_estimators=nestimators)
        Forest.fit(X_dense, y_binary)
        temp = pd.Series(Forest.feature_importances_, index=terms)
        temp.sort(ascending=False)
        temp[:7].plot(kind="bar", figsize=(8,3), fontsize=14, grid=False, alpha=0.7, linewidth=0.0)
        xticks(rotation=50, ha="right")
        title("Group %d  (n = %d)" % (k, obs_per_cluster[k]), fontsize=20)
        ylabel('Relative Feature Importance')
        show()
```
